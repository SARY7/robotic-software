```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# import images
# %matplotlib inline
rgb_image = mpimg.imread('importing_data/cup_rgb.png')
depth_map = mpimg.imread('importing_data/cup_depth.png')

plt.subplot(1,2,1)
plt.imshow(rgb_image)
plt.title("RGB image")

plt.subplot(1,2,2)
plt.imshow(depth_map)
plt.title("depth map")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
rows, cols, data  = depth_map.shape
count = 0
pixel_cut = 3
for x in range(1, 250):
    for y in range(1, 180):
        if (x % pixel_cut == 0 and y % pixel_cut == 0):
            # Count points done
            count+=1
            pixel_color = (rgb_image[y, x, 0], rgb_image[y, x, 1], rgb_image[y, x, 2])
            depth = depth_map[y, x]
       
            ax.scatter(x, depth, y, c = pixel_color)
plt.gca().invert_zaxis()

elevation = 30 # Up/Down
azimuth = 300 # Left/Right
plt.show()
```
